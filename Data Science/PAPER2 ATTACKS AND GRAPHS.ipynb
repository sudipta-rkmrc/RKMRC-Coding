# 10) NPCR Code

Explanation:
# NPCR Definition: 
NPCR measures the percentage of pixels that differ between two images.
# Same Images: 
If the images are identical, there are no pixel differences between them.
# Calculation: 
Since there are no differing pixels, the numerator in the NPCR formula (number of differing pixels) is 0.

```python
from PIL import Image
import numpy as np

def calculate_npcr(image1, image2):
    # Load images
    #image1 = Image.open(image1_path).convert('L')  # Convert to grayscale
    #image2 = Image.open(image2_path).convert('L')  # Convert to grayscale

    # Convert images to numpy arrays
    img1_array = np.array(image1)
    img2_array = np.array(image2)

    # Check if images have the same shape
    if img1_array.shape != img2_array.shape:
        raise ValueError("Images must have the same dimensions")

    # Calculate the difference matrix
    diff = img1_array != img2_array

    # Count the number of differing pixels
    num_diff_pixels = np.sum(diff)

    # Total number of pixels
    total_pixels = img1_array.size

    # Calculate NPCR
    npcr = (num_diff_pixels / total_pixels) * 100

    return npcr
```

# 11) UACI Code

# Key Points:
# a) Normalization: 
The absolute difference between corresponding pixels in the two images is normalized by dividing by 255 (since the images are assumed to be 8-bit grayscale).
# b)Averaging:
The mean of these normalized differences is calculated, and the result is multiplied by 100 to express it as a percentage.
# c)Grayscale Conversion:
Images are converted to grayscale to simplify the comparison process.
# Interpretation:
# a)Higher UACI Values: 
Indicate that the average change in intensity between corresponding pixels is significant, suggesting that the images are quite different.
# b) Lower UACI Values: 
Suggest that the images have similar intensity levels at corresponding pixels.

```python
from PIL import Image
import numpy as np

def calculate_uaci(image1, image2):
    # Load images
    #image1 = Image.open(image1_path).convert('L')  # Convert to grayscale
    #image2 = Image.open(image2_path).convert('L')  # Convert to grayscale

    # Convert images to numpy arrays
    img1_array = np.array(image1, dtype=np.float32)
    img2_array = np.array(image2, dtype=np.float32)

    # Check if images have the same shape
    if img1_array.shape != img2_array.shape:
        raise ValueError("Images must have the same dimensions")

    # Calculate UACI
    diff_intensity = np.abs(img1_array - img2_array)  # Absolute difference
    uaci = np.mean(diff_intensity / 255) * 100  # Normalize and average

    return uaci

```

```python
# Example usage:
image1_path = '../Results/SIPI Result Integer/Stego/wtr/watermarkairplane_splash.tiff'
image2_path = '../Results/SIPI Result Integer/Watermark/splash.tiff'
im1=cv2.imread(image1_path,0)
im2=cv2.imread(image2_path,0)
im2=cv2.resize(im2,(32,32))

uaci_value = calculate_uaci(im1, im2)
print(f"UACI between the two images: {uaci_value:.2f}%")
```

# 12) Bit Error Rate (BER) code

To calculate the Bit Error Rate (BER) between two images, you need to compare their binary representations bit by bit. BER will measure the proportion of bits that are different between the two images, making it an indicator of how much they differ at the binary level. This approach is useful in image transmission systems, where bit errors can significantly affect image quality.

Steps to Calculate BER Between Two Images
Load the Images: Convert images to grayscale (or handle color images if needed) and ensure they are of the same dimensions.
Convert to Binary Format: Convert the pixel values of each image into their binary representations.
Compare Bits: Compare the binary bits of the two images to count the number of differing bits.
Calculate BER: Divide the number of differing bits by the total number of bits.

```python
from PIL import Image
import numpy as np

def image_to_bitstream(image):
    """
    Convert an image to a bit stream (binary representation).

    :param image_path: Path to the image file
    :return: A list of bits representing the image
    """
    # Load the image and convert it to grayscale
    #image = Image.open(image_path).convert('L')
    
    # Convert image to numpy array of pixel values
    image_array = np.array(image)
    
    # Flatten the array to 1D
    flat_pixels = image_array.flatten()
    
    # Convert each pixel value to its binary representation and concatenate bits
    bitstream = ''.join([f'{pixel:08b}' for pixel in flat_pixels])  # 8-bit binary for grayscale
    
    return bitstream

def calculate_ber(bitstream1, bitstream2):
    """
    Calculate Bit Error Rate (BER) between two bit streams.

    :param bitstream1: First bit stream (string of '0's and '1's)
    :param bitstream2: Second bit stream (string of '0's and '1's)
    :return: Bit Error Rate (BER)
    """
    # Ensure both bitstreams are the same length
    if len(bitstream1) != len(bitstream2):
        raise ValueError("Bitstreams must be of the same length")
    
    # Calculate the number of differing bits
    errors = sum(b1 != b2 for b1, b2 in zip(bitstream1, bitstream2))
    
    # Calculate BER
    ber = errors / len(bitstream1)
    
    return ber

```

```python
image1_path = '../Results/SIPI Result Integer/Stego/wtr/watermarkairplane_splash.tiff'
image2_path = '../Results/SIPI Result Integer/Watermark/splash.tiff'
im1=cv2.imread(image1_path,0)
im2=cv2.imread(image2_path,0)
im2=cv2.resize(im2,(32,32))
# Convert images to bit streams
bitstream1 = image_to_bitstream(im1)
bitstream2 = image_to_bitstream(im2)

# Calculate BER
ber_value = calculate_ber(bitstream1, bitstream2)
print(f"Bit Error Rate (BER) between the two images: {ber_value:.6f}")
```

# 13) Chi Squared image comparison

```python
from PIL import Image
import numpy as np
from scipy.stats import chi2_contingency

def chi_square_image_comparison(image1, image2, bins=256):
    # Load and convert images to grayscale
    #image1 = Image.open(image1_path).convert('L')
    #image2 = Image.open(image2_path).convert('L')

    # Convert images to numpy arrays
    img1_array = np.array(image1)
    img2_array = np.array(image2)

    # Compute histograms (categorize pixel intensities)
    hist1, _ = np.histogram(img1_array, bins=bins, range=(0, 256))
    hist2, _ = np.histogram(img2_array, bins=bins, range=(0, 256))

    # Create a contingency table from the histograms
    contingency_table = np.array([hist1, hist2])
    #print(contingency_table)
    # Perform the Chi-square test
    chi2, p, dof, expected = chi2_contingency(contingency_table)

    print(f"Chi-square statistic: {chi2}")
    print(f"p-value: {p}")
    print(f"Degrees of freedom: {dof}")
    print("Expected frequencies:\n", expected)
```

```python
# Example usage:
image1_path = '../Results/SIPI Result Integer/Cover/airplane.tiff'
image2_path = '../Results/SIPI Result Integer/Cover/lena.tiff'
im1=cv2.imread(image1_path,0)
im2=cv2.imread(image2_path,0)

chi_square_image_comparison(im1, im2)
```
