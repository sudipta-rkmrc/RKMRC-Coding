# 3. Create a black window with three color channels with resolution 512 x 512. 
Then create three ‘B’ ‘G’ ‘R’ trackbars using predefined functions of 
OpenCV library. Set the values of channels from 0 to 255. Merging th 
black window with these color trackbars.

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
canvas = np.zeros((300, 500, 3), dtype=np.uint8)
cv2.namedWindow('Color Mixer')
```

```python
def update_color(dummy=None):
   
    blue = cv2.getTrackbarPos('Blue', 'Color Mixer')
    green = cv2.getTrackbarPos('Green', 'Color Mixer')
    red = cv2.getTrackbarPos('Red', 'Color Mixer')

    canvas[:] = [blue, green, red]

    cv2.imshow('Color Mixer', canvas)
```

```python
cv2.createTrackbar('Blue', 'Color Mixer', 0, 255, update_color)
cv2.createTrackbar('Green', 'Color Mixer', 0, 255, update_color)
cv2.createTrackbar('Red', 'Color Mixer', 0, 255, update_color)
```

```python
update_color()
while True:
    key = cv2.waitKey(1) & 0xFF
    if key == ord('q'):
        break

cv2.destroyAllWindows()
```
