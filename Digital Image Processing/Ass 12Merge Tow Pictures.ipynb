# 12. Take a collection of images and make a collage will smaller version of it.

```python
import cv2
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
def merge_and_resize_images(image1_path, image2_path, output_size):
    image1 = cv2.imread(image1_path)
    image2 = cv2.imread(image2_path)

    image1_resized = cv2.resize(image1, output_size)
    image2_resized = cv2.resize(image2, output_size)

    merged_image = cv2.hconcat([image1_resized, image2_resized])

    return merged_image
```

```python
image1_path = 'lena.tiff'
image2_path = 'elaine.tiff'
output_size = (300, 300) 

merged = merge_and_resize_images(image1_path, image2_path, output_size)
```

```python
plt.imshow(merged)
plt.show()
```
