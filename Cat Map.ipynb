```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def arnold_cat_map(img, n):
    r, c = img.shape
    for _ in range(n):
        img_new = np.zeros_like(img)  
        for x in range(r):
            for y in range(c):
                x_new = (2*x + y) % r
                y_new = (x + y) % r
                img_new[x_new, y_new] = img[x, y]
        
        img = img_new  
    return img
```

```python
img = cv2.imread('fishingboat.tiff',0)
img = cv2.resize(img, (101,101))
for i in range(30):
    plt.title("Iteration "+str(i))
    plt.imshow(arnold_cat_map(img,i),cmap='gray')
    plt.show()
```
